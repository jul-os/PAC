# Объектно-ориентированное программирование в Python

## Базовые принципы ООП
* Инкапсуляция — размещение одного объекта или класса внутри другого для разграничения доступа к ним
* Наследование — способность объекта или класса базироваться на другом объекте или классе. Это главный механизм для повторного использования кода. Наследственное отношение классов четко определяет их иерархию
* Полиморфизм — реализация задач одной и той же идеи разными способами
* Абстракция — отделение концепции от ее экземпляра

### Инкапсуляция

Инкапсуляция есть объединение в единое целое данных и алгоритмов обработки этих данных. В рамках ООП данные называются полями объекта, а алгоритмы - объектными методами.

Инкапсуляция позволяет в максимальной степени изолировать объект от внешнего окружения. Она существенно повышает надежность разрабатываемых программ, т.к. локализованные в объекте алгоритмы обмениваются с программой сравнительно небольшими объемами данных, причем количество и тип этих данных обычно тщательно контролируются. В результате замена или модификация алгоритмов и данных, инкапсулированных в объект, как правило, не влечет за собой плохо прослеживаемых последствий для программы в целом (в целях повышения защищенности программ в ООП почти не используются глобальные переменные).

### Наследование

Наследование есть свойство объектов порождать своих потомков. Объект-потомок автоматически наследует от родителя все поля и методы, может дополнять объекты новыми полями и заменять (перекрывать) методы родителя или дополнять их.

Принцип наследования решает проблему модификации свойств объекта и придает ООП в целом исключительную гибкость. При работе с объектами программист обычно подбирает объект, наиболее близкий по своим свойствам для решения конкретной задачи, и создает одного или нескольких потомков от него, которые «умеют» делать то, что не реализовано в родителе.

Последовательное проведение в жизнь принципа «наследуй и изменяй» хорошо согласуется с поэтапным подходом к разработке крупных программных проектов и во многом стимулирует такой подход.

### Полиморфизм

Полиморфизм - это свойство родственных объектов (т.е. объектов, имеющих одного общего родителя) решать схожие по смыслу проблемы разными способами. В рамках ООП поведенческие свойства объекта определяются набором входящих в него методов. Изменяя алгоритм того или иного метода в потомках объекта, программист может придавать этим потомкам отсутствующие у родителя специфические свойства. Для изменения метода необходимо перекрыть его в потомке, т.е. объявить в потомке одноименный метод и реализовать в нем нужные действия. В результате в объекте-родителе и объекте-потомке будут действовать два одноименных метода, имеющие разную алгоритмическую основу и, следовательно, придающие объектам разные свойства.

### Абстракция

Абстракция в объектно-ориентированном программировании — это использование только тех характеристик объекта, которые с достаточной точностью представляют его в данной системе. Основная идея состоит в том, чтобы представить объект минимальным набором полей и методов и при этом с достаточной точностью для решаемой задачи.

Абстракция является основой объектно-ориентированного программирования и позволяет работать с объектами, не вдаваясь в особенности их реализации.

# Создание класса

````python
class MyClass:
    val = 2  # Свойство класса

    def mult(self, x):  # Методы класса
        return self.val * x
    
    def power(self, x):
        return x ** self.val
````

# Инкапсуляция

Атрибут может быть объявлен приватным (внутренним) с помощью нижнего подчеркивания перед именем, но настоящего скрытия на самом деле не происходит – все на уровне соглашений.

````python
class MyClass:
    def __init__(self):
        self._val = 3  # Приватное свойство класса
        self.__priv_val = 8  # Очень приватное свойство класса
        
    def _factorial(self, x): # Приватный метод класса
        fact = 1
        for i in range(1, x):
            fact *= i
        return fact

    def mult(self, x):  # Метод класса
        return self._val * self._factorial(x)
````

### Работа с приватными данными

````python
class Item:
    def __init__(self, count=3, max_count=16):
        self._count = count
        self._max_count = 16
        
    def update_count(self, val):
        if val <= self._max_count:
            self._count = val
            return True
        else:
            return False
        
    # Свойство объекта. Не принимает параметров кроме self, вызывается без круглых скобок
    # Определяется с помощью декоратора property
    @property
    def count(self):
        return self._count
    
    
    # Ещё один способ изменить атрибут класса
    @count.setter
    def count(self, val):
        self._count = val
        if val <= self._max_count:
            self._count = val
        else:
            pass
    
    @staticmethod
    def static():
        print('I am function')
    
    @classmethod
    def my_name(cls):
        return cls.__name__
````

````python
item = Item(2)
print(item.count)
ret = item.update_count(8)
print(ret, item.count)
````

## Множественное наследование

````python
class Fruit(Item):
    def __init__(self, ripe=True, **kwargs):
        super().__init__(**kwargs)
        self._ripe = ripe


class Food(Item):
    def __init__(self, saturation, **kwargs):
        super().__init__(**kwargs)
        self._saturation = saturation
        
    @property
    def eatable(self):
        return self._saturation > 0
````

````python
class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color
    
    @property
    def color(self):
        return self._color
````

````python
apple = Apple(False, color='идг')
print(apple.count)
print(apple.color)
print(apple.eatable)
#print(apple.saturation) or ripe нет потому что оба этих приватные
````

# Полиморфизм

````python
class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color
    
    @property
    def color(self):
        return self._color
        
    @property
    def eatable(self):
        """
        Переопределённая функция класса Food. Добавление проверки на спелость
        """
        return super().eatable and self._ripe
````

apple.eatable → Apple.eatable → Food.eatable + проверка на ripe

Пошагово:

    super().eatable - вызывает свойство eatable из родительского класса Food

    Food.eatable - возвращает self._saturation > 0 (проверка на сытость)

    and self._ripe - добавляет проверку, что фрукт должен быть спелым

````python
apple = Apple(False, color='green')
print(apple.count)
print(apple.color)
print(apple.eatable)
````

# Специальные методы

````python
# Магические методы python
class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color
    
    @property
    def color(self):
        return self._color
        
    @property
    def eatable(self):
        return super().eatable and self._ripe
    
    def __call__(self):
        """ Вызов как функции """
        if self.eatable:
            new_count = max(self.count - 1, 0)
            self.update_count(new_count)            
    
    def __str__(self):
        """ Вызов как строки """
        return f'Stack of {self.count} {self.color} apples' 
            
    def __len__(self):
        """ Получение длины объекта """
        return self.count
````

### Имитация списка

````python
class MultLists:
    def __init__(self, lst1, lst2):
        self._lst1 = lst1
        self._lst2 = lst2
        
    def __getitem__(self, index):
        """ Получение элемента по индексу """
        if index > len(self):
            raise IndexError(f'Index {index} more then {len(self)}')
        return self._lst1[index], self._lst2[index]
        
    def __len__(self):
        return min(len(self._lst1), len(self._lst2))
````

````python
mlst = MultLists([1, 4, 5, 7], ['a', 'b', 'c'])
print(len(mlst))
print(mlst[2])
print(mlst[5])
````

### Числовые операции

````python
# (задание 2)
class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color='green', saturation=10):
        super().__init__(saturation=saturation, ripe=ripe, count=count, max_count=max_count)
        self._color = color
    
    @property
    def color(self):
        return self._color
        
    @property
    def eatable(self):
        return super().eatable and self._ripe
    
    def __add__(self, num):
        """ Сложение с числом """
        return self.count + num
    
    def __mul__(self, num):
        """ Умножение на число """
        return self.count * num
    
    def __lt__(self, num):
        """ Сравнение меньше """
        return self.count < num
    
    def __len__(self):
        """ Получение длины объекта """
        return self.count
````

````python
# Перенести все операции по работе с количеством объектов из класса Apple в класс Item

class Item:
    def __init__(self, count=3, max_count=16):
        self._count = count
        self._max_count = 16

    def update_count(self, val):
        if 0 <= val <= self._max_count:
            self._count = val
            return True
        else:
            return False

    @property
    def count(self):
        return self._count

    @count.setter
    def count(self, val):
        self._count = val
        if 0 <= val <= self._max_count:
            self._count = val
        else:
            pass

    @staticmethod
    def static():
        print("I am function")

    @classmethod
    def my_name(cls):
        return cls.__name__

    def __len__(self):
        """Получение длины объекта"""
        return self.count

    def __add__(self, num):
        """Сложение с числом"""
        return self.count + num

    def __mul__(self, num):
        """Умножение на число"""
        return self.count * num

    def __lt__(self, num):
        """Сравнение меньше"""
        return self.count < num

    """2. Дополнить класс Item остальными опрерациями сравнения (>, <=, >=, ==), вычитания, 
а также выполнение операций +=, *=, -=. Все изменения количества должны быть в переделах 0, max_count"""

    def __gt__(self, num):
        """Сравнние больше"""
        return self.count > num
    def __ge__(self, num):
        return self.count >= num
    def __le__(self, num):
        return self.count <= num
    def __eq__(self, num):
        return self.count == num
    def __sub__(self, num):
        return self.count - num
    def __iadd__(self, num):
        _new_count = self.count + num
        _new_count = max(0, min(_new_count, self._max_count))
        self.count = _new_count
        return self.count
    def __isub__(self, num):
        _new_count = self.count - num
        _new_count = max(0, min(_new_count, self._max_count))
        self.count = _new_count
        return self.count
    def __imul__(self, num):
        _new_count = self.count * num
        _new_count = max(0, min(_new_count, self._max_count))
        self.count = _new_count
        return self.count


# тогда класс apple выглядит так


class Apple(Fruit, Food):
    def __init__(self, ripe, count=1, max_count=32, color="green", saturation=10):
        super().__init__(
            saturation=saturation, ripe=ripe, count=count, max_count=max_count
        )
        self._color = color

    @property
    def color(self):
        return self._color

    @property
    def eatable(self):
        """
        Переопределённая функция класса Food. Добавление проверки на спелость
        """
        return super().eatable and self._ripe

    def __call__(self):
        """Вызов как функции"""
        if self.eatable:
            new_count = max(self.count - 1, 0)
            self.update_count(new_count)

    def __str__(self):
        """Вызов как строки"""
        return f"Stack of {self.count} {self.color} apples"
````

4. Создать класс Inventory, который содержит в себе список фиксированной длины. Заполнить его None. Доступ в ячейку осуществляется по индексу.   
    4.1 Добавить возможность добавлять в него съедобные объекты в определённые ячейки.  
    4.2 Добавить возможность уменьшать количество объектов в списке.  
    4.3 При достижении нуля, объект удаляется из инвентаря.  

````python
class Inventory:
    def __init__(self, size=10):
        """Инициализация инвентаря фиксированного размера"""
        self._size = size
        self._slots = [None] * size  # Заполняем None

    def __len__(self):
        """Возвращает размер инвентаря"""
        return self._size

    def __getitem__(self, index):
        """Доступ к ячейке по индексу"""
        if 0 <= index < self._size:
            return self._slots[index]
        else:
            raise IndexError(f"Индекс {index} вне диапазона [0, {self._size-1}]")

    def __setitem__(self, index, value):
        """Установка значения в ячейку по индексу"""
        if 0 <= index < self._size:
            self._slots[index] = value
        else:
            raise IndexError(f"Индекс {index} вне диапазона [0, {self._size-1}]")

    def __str__(self):
        """Строковое представление инвентаря"""
        result = "Инвентарь:\n"
        for i, item in enumerate(self._slots):
            result += f"{i}: {item}\n"
        return result

    def add_item(self, index, item):
        """
        4.1 Добавление съедобного объекта в определённую ячейку
        """
        if not isinstance(item, Food):
            raise ValueError(
                "Можно добавлять только съедобные объекты (наследники Food)"
            )

        if 0 <= index < self._size:
            if self._slots[index] is None:
                self._slots[index] = item
                return True
            else:
                # Можно добавить логику объединения стеков, если это один тип предмета
                print(f"Ячейка {index} уже занята")
                return False
        else:
            raise IndexError(f"Индекс {index} вне диапазона [0, {self._size-1}]")

    def decrease_item(self, index, amount=1):
        """
        4.2 Уменьшение количества объекта в ячейке
        4.3 Удаление объекта при достижении нуля
        """
        if 0 <= index < self._size:
            item = self._slots[index]

            if item is None:
                print(f"Ячейка {index} пуста")
                return False

            if not isinstance(item, Item):
                print(f"Объект в ячейке {index} не поддерживает операции с количеством")
                return False

            # Уменьшаем количество
            if hasattr(item, "count"):
                item.count -= amount

                # Проверяем, достигло ли количество нуля
                if item.count <= 0:
                    self._slots[index] = None  # 4.3 Удаляем объект
                    print(f"Объект в ячейке {index} удалён (количество достигло нуля)")
                    return True
                else:
                    print(f"Количество в ячейке {index} уменьшено до {item.count}")
                    return True
            else:
                print(f"Объект в ячейке {index} не имеет свойства count")
                return False
        else:
            raise IndexError(f"Индекс {index} вне диапазона [0, {self._size-1}]")

    def use_item(self, index):
        """Использование предмета (альтернативный способ уменьшения)"""
        if 0 <= index < self._size:
            item = self._slots[index]

            if item is None:
                print(f"Ячейка {index} пуста")
                return False

            if hasattr(item, "__call__"):
                item()  # Вызываем объект как функцию

                # Проверяем количество после использования
                if hasattr(item, "count") and item.count <= 0:
                    self._slots[index] = None
                    print(f"Объект в ячейке {index} удалён после использования")
                return True
            else:
                print(f"Объект в ячейке {index} нельзя использовать")
                return False
        else:
            raise IndexError(f"Индекс {index} вне диапазона [0, {self._size-1}]")

    def find_empty_slot(self):
        """Поиск первой пустой ячейки"""
        for i in range(self._size):
            if self._slots[i] is None:
                return i
        return -1  # Все ячейки заняты

    def get_items_info(self):
        """Получение информации о всех предметах в инвентаре"""
        info = []
        for i, item in enumerate(self._slots):
            if item is not None:
                if hasattr(item, "count"):
                    info.append(f"{i}: {item} (количество: {item.count})")
                else:
                    info.append(f"{i}: {item}")
            else:
                info.append(f"{i}: Пусто")
        return info
````

````python
# Создаем инвентарь
inventory = Inventory(size=5)

# Создаем предметы
apple = Apple(ripe=True, count=3, color="red")
grape = Grape(ripe=True, count=5, color="purple")
cake = Cake(count=2, color="chocolate")

# 4.1 Добавляем предметы в инвентарь
inventory.add_item(0, apple)
inventory.add_item(2, grape)
inventory.add_item(3, cake)

print("После добавления:")
for info in inventory.get_items_info():
    print(info)

# 4.2 Уменьшаем количество
print("\nУменьшаем количество яблок:")
inventory.decrease_item(0, 2)  # Уменьшаем на 2
inventory.decrease_item(0, 2)  # Уменьшаем ещё на 2 (должен удалиться)

print("\nПосле уменьшения:")
for info in inventory.get_items_info():
    print(info)

print("\nИспользуем виноград:")
inventory.use_item(2)  # Используем 1 виноградинку

print("\nФинальное состояние инвентаря:")
print(inventory)
````
